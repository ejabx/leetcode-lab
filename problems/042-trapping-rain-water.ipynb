# [Trapping Rain Water](https://leetcode.com/problems/trapping-rain-water/description/)

Given height[] where each element is a bar’s height, compute how much rainwater is trapped between them.

## Strategy

### Two-pointer approach:

1. Track left and right pointers from both ends.
2. Track leftMax and rightMax as you go.
3. If height[left] < height[right], then:
4a. If height[left] < leftMax, water can be trapped.
4b. Else, update leftMax.
5. Move the shorter side pointer inward.

### Alternative: Precompute max arrays

1. Build leftMax[] and rightMax[] arrays.
2. Water trapped at i is min(leftMax[i], rightMax[i]) - height[i].

__Time__: O(n) — one pass or two for preprocessing  
__Space__: O(1) (if using two-pointer), O(n) with precomputed arrays.
