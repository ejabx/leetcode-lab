## Bonus: Flashcard Topics

### Two Pointers

**Q:** When is the two-pointer technique most effective?  
**A:** When working with sorted arrays or problems that involve finding pairs, triplets, or quadruplets with a target sum. Useful for reducing O(nÂ²) to O(n).

**Q:** How do you avoid duplicates in 3Sum or 4Sum?  
**A:** Sort the array, and skip repeated elements using conditions like `if i > 0 and nums[i] == nums[i-1]: continue`.

**Q:** Whatâ€™s the strategy behind the Container With Most Water problem?  
**A:** Use two pointers at both ends and move the pointer pointing to the shorter line inward.

### Sliding Window

**Q:** What is the basic sliding window pattern?  
**A:** Expand the window by moving the right pointer, then shrink it from the left when the window becomes invalid.

**Q:** How do you track character frequencies in string sliding window problems?  
**A:** Use a `Counter` or dictionary to store character counts.

**Q:** Which data structure optimizes sliding window maximum?  
**A:** A monotonic deque (stores indices in decreasing value order).

### Prefix/Suffix Techniques

**Q:** What kind of problems are prefix/suffix arrays good for?  
**A:** Problems that require computing results over a range or excluding the current element, like "Product of Array Except Self".

**Q:** How is prefix/suffix used in Trapping Rain Water?  
**A:** Precompute `leftMax` and `rightMax` arrays to find the min boundary height at each index.

### Monotonic Stack

**Q:** What type of problems use monotonic stacks?  
**A:** Problems involving the next greater or smaller element, histogram max area, and element ordering with constraints.

**Q:** What is the monotonic stack invariant for Daily Temperatures?  
**A:** Maintain a stack of indices where temperatures are in decreasing order. Pop when the current temp is higher.

**Q:** How does `Remove Duplicate Letters` use a stack?  
**A:** It maintains a stack for result chars, a set for `seen`, and a map for remaining character counts. Pop from stack if a smaller character appears and the top can still appear later.

### Greedy Techniques

**Q:** When is greedy a good strategy?  
**A:** When local optimal choices lead to a global optimal solution â€” often in problems involving max area, min length, or lexicographic order.

**Q:** What greedy rule is applied in `Remove Duplicate Letters`?  
**A:** Always keep the smallest lexicographical character while ensuring all characters still appear later.

### Bonus: Concept Checks

**Q:** How do you decide between prefix sums and sliding window?  
**A:** Use prefix sums when you're making repeated range queries; use sliding window when working with dynamic subarrays under constraints.

**Q:** Whatâ€™s a common pattern in monotonic stack problems?
